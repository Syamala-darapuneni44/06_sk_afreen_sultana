```python
# Colabâ€friendly version with a valid Groq model

from google.colab import userdata

GROQ_API_KEY = userdata.get("GROQ_API_KEY")
if not GROQ_API_KEY:
    raise RuntimeError("GROQ_API_KEY not found in Colab Secrets. Please add it first.")

from langchain_groq import ChatGroq
from langchain_community.tools import Tool
from langchain.agents import initialize_agent

llm = ChatGroq(
    model="llama-3.3-70b-versatile",   # <-- use a valid, active model
    temperature=0.7,
    groq_api_key=GROQ_API_KEY
)

def recipe_generator(ingredients: str) -> str:
    prompt = f"Generate a detailed, step-by-step food recipe using these ingredients: {ingredients}. Include cooking time and serving size."
    response = llm.invoke(prompt)
    return response.content

tools = [
    Tool(
        name="Recipe Generator",
        func=recipe_generator,
        description="Use this tool to generate recipes based on given ingredients."
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True
)

query = "I have chicken, onions, tomatoes, and garlic. Suggest a recipe."

try:
    result = agent.invoke(query)
    # Depending on version, result might be a dict or object
    if hasattr(result, "content"):
        print("ğŸ´ Recipe Suggestion:\n", result.content)
    else:
        # if dict
        print("ğŸ´ Recipe Suggestion:\n", result.get("output") or result.get("text") or result)
except Exception as e:
    print("âŒ Error occurred:", e)
```
